```python
import pandas as pd 
import numpy as np
first = pd.Series(np.random.randint(1,4 ,size=(100)),name='first')
second = pd.Series(np.random.randint(1,3 ,size=(100)), name= 'second')
third = pd.Series(np.random.randint( 10000,30000 ,size=(100)),name= 'Third')
print( third)


# In[13]:


data= pd.concat([first, second, third], axis=1)
df = pd.DataFrame(data)
print(df)


# In[15]:


df.rename(columns={'first':'bedrs','second':'bathrs','Third':'price_sqr_meter'},inplace = True)
print(df)


# In[23]:


bigcolumn = df['bedrs'].append(df['bathrs'].append(df['price_sqr_meter'])).reset_index(drop=True)
print(bigcolumn)


# Ques: Ops it seems it is going only until index 99. Is it true?
# Ans:  yes it is True.

# In[27]:


bigcolumn.reset_index(inplace=True, drop=True)
print(bigcolumn)


# In[ ]:




```
